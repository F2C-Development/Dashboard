```python
import pandas as pd
import numpy as np
```

```python
siconfi_df=pd.read_csv('../siconfi.txt')

mask=siconfi_df['Município [-]']=='Barão de Monte Alto'
siconfi_df.loc[mask,'Município [-]']='Barão do Monte Alto'
mask=siconfi_df['Município [-]']=='Gracho Cardoso'
siconfi_df.loc[mask,'Município [-]']='Graccho Cardoso'
siconfi_df.to_csv('../siconfi.txt', index=False)

#quando for colocar na base - lembrar de tirar as três lá de baixo

siconfi_df
```

```python
ibge_df=pd.read_csv('../ibge.txt')

ibge_df
```

```python


ibge_df=ibge_df.fillna(np.nan)
ibge_df = ibge_df.replace("nan", np.nan)
ibge_df.to_csv('../ibge.txt',index=False)
ibge_df.count()
```

```python
df=ibge_df.merge(siconfi_df, how='inner')
df
```

```python
mask=df['DC [2024]'].isna()
df[mask]
```
